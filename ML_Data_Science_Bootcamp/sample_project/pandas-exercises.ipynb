```python
#Import the car sales DataFrame with missing data
cars_data_missing = pd.read_csv("car-sales-missing-data.csv")
cars_data_missing
```

```python
#Fill the Odometer column missing values with the mean of the column inplace
cars_data_missing["Odometer"].fillna(cars_data_missing["Odometer"].mean(numeric_only=True), inplace=True)
cars_data_missing
```

```python
#Remove the rest of the missing data inplace
cars_data_missing.dropna(inplace=True)
cars_data_missing
```

```python
#Create a seats column where every row has a value of 5
cars_data["Seats"] = 5
cars_data
```

```python
#Create a column called "Engine Size" with random values between 1.3 and 4.5
import random
cars_data["Engine Size"] = random.uniform(1.3, 4.5)
cars_data
```

```python
#Create a column which represents the price of a car per kilometer
cars_data["Price per kilometer"] = random.uniform(4.5, 7.0)
cars_data
```

```python
#Remove the last column added using .drop()
cars_data.drop("Price per kilometer", axis=1, inplace=True)
cars_data
```

```python
#Shuffle the DataFrame using sample() with the frc parameter set to 1 
#Safe the shuffled DataFrame to new one
shuffled_frame = cars_data.sample(frac=1)
shuffled_frame
```

```python
#reset the indexes of the shuffled DataFrame
shuffled_frame.reset_index(inplace=True)
shuffled_frame
```

```python
#Change the Odometer values from kilometers to miles using a Lambda function
cars_data["Odometer (KM)"] = cars_data["Odometer (KM)"].apply(lambda x: x/1.6)
cars_data
```

```python
#Change the title of the Odometer (KM) to represent miles instead of kilometers
cars_data = cars_data.rename(columns={"Odometer (KM)": "Odometer (Miles)"})
cars_data
```
